```python
import numpy as np  
import matplotlib.pyplot as plt  
get_ipython().run_line_magic('matplotlib', 'inline')
import pandas as pd  
from apyori import apriori
import itertools
```

```python
df_mlt = pd.read_csv('/Users/mtran/Documents/SMUDataSciences/2019-Summer/Diabetic_Cleaned.csv')
print("Finished loading data")
```

```python
#make dataframes for each column
ordinal = ["age"]
categorical = ["gender", "max_glu_serum", "A1Cresult","diabetesMed"]
medicines = ["insulin"]
```

```python
df_mlt_ordinal = df_mlt[ordinal]
df_mlt_categorical = df_mlt[categorical]
df_mlt_medicines = df_mlt[medicines]
```

```python
#Make age ordinal
for i in range(0, 10):
    val = "[" + str(i*10) + "-" + str((i+1)*10) + ")"
    df_mlt_ordinal.loc[df_mlt_ordinal['age'] == val, 'age'] = i
```

```python
#Creating dataframe
df_mlt_med_x = pd.concat([df_mlt_ordinal, df_mlt_categorical, df_mlt_medicines], axis = 1)
print("Done")
```

```python
%%time

records = []  
for i in range(0, 10000):  
    records.append([str(df_mlt_med_x.values[i,j]) for j in range(0, 6)]) 
```

```python
%%time

association_rules = apriori(records, min_support=0.005, min_confidence=0.02, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
def gen():
    for i in range(10):
        yield i
```

```python
class LenGen(object):
    def __init__(self,gen,length):
        self.gen=gen
        self.length=length
    def __call__(self):
        return itertools.islice(self.gen(),self.length)
    def __len__(self):
        return self.length

lgen=LenGen(gen,10)
```

```python
def new_gen():
    for i in lgen():
        yield float(i)/len(lgen)

for i in new_gen():
    print(i)
```
